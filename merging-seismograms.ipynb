```python
import obspy
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
# matplotlib.use('Agg')
```

```python
# read and merge multiple streams
stream = obspy.read("https://examples.obspy.org/dis.G.SCZ.__.BHE")
stream += obspy.read("https://examples.obspy.org/dis.G.SCZ.__.BHE.1")
stream += obspy.read("https://examples.obspy.org/dis.G.SCZ.__.BHE.2")
```

```python
stream.sort(['starttime'])
datetime = stream[0].stats.starttime.timestamp
```

```python
ax = plt.subplot(4, 1, 1)
for i in range(3):
    # plt.subplot(4, 1, i + 1, sharex=ax)
    plt.plot(stream[i].times(), stream[i].data)

```

```python
stream.merge(method=1)
plt.subplot(4, 1, 4, sharex=ax)
plt.plot(stream[0].times(), stream[0].data, 'r')
plt.show()
```
